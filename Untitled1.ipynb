### What This Notebook Does

SustAInable predicts which ZIP codes will see **elevated heat illness** during an approaching extreme heat event so cities can deploy cooling resources before people need to be hospitalized. This notebook explores and validates the **structural vulnerability features** that feed the model:
| Layer | Source | What it measures |
|-------|--------|	|
| Health burden | CDC PLACES 2024 | Chronic disease prevalence by ZIP code |
| Socioeconomic vulnerability | ACS 5-Year Estimates 2022 | Poverty, housing age, elderly %, disability rate, renter rate |

**Outcome variable note:** The label — elevated heat illness ED visit rate from CDC NSSP - requires a CDC data use agreement and is not included here. This notebook establishes and validates the feature layer, and builds a composite structural vulnerability index as a baseline comparator to the CDC Heat and Health Index (HHI).

### Demonstrating Point-in-Polygon Join with GeoPandas

To perform a point-in-polygon join, you typically need two GeoDataFrames:
1.  A GeoDataFrame containing **points** (e.g., locations of interest, facilities, etc.).
2.  A GeoDataFrame containing **polygons** (e.g., administrative boundaries like census tracts or ZCTAs).

The `geopandas.sjoin` function allows you to join these based on their spatial relationship. For a 'point-in-polygon' join, we use `op='within'` to find which points fall within which polygons.

Since we already have `philly_heat_gdf` (Philadelphia census tracts as polygons), we'll generate some dummy point data for demonstration purposes.

````python
import geopandas as gpd
from shapely.geometry import Point
import numpy as np

# Assuming philly_heat_gdf is already loaded and contains the census tract polygons

# 1. Create some dummy point data within the approximate bounds of Philadelphia
#    We'll use the bounding box of philly_heat_gdf to ensure points are relevant.
minx, miny, maxx, maxy = philly_heat_gdf.total_bounds

# Generate 100 random points within Philadelphia's bounding box
np.random.seed(42) # for reproducibility
N_POINTS = 100
sample_points = gpd.GeoDataFrame(
    {
        'point_id': range(N_POINTS),
        'data_value': np.random.rand(N_POINTS) * 100 # Dummy data for points
    },
    geometry=gpd.points_from_xy(
        np.random.uniform(minx, maxx, N_POINTS),
        np.random.uniform(miny, maxy, N_POINTS)
    ),
    crs=philly_heat_gdf.crs # Ensure CRS matches the polygons
)

print("Generated sample points (first 5 rows):")
display(sample_points.head())

# 2. Perform the point-in-polygon spatial join
#    'left_factor' means keep all points and add polygon attributes if a match is found
#    'within' operator checks if a point is within a polygon

points_in_tracts = gpd.sjoin(
    sample_points,
    philly_heat_gdf[['geoid10', 'hvi_score', 'geometry']], # Select relevant columns from polygons
    how='left',
    predicate='within' # Changed 'op' to 'predicate'
)

print("\nResults of point-in-polygon join (first 5 rows):")
display(points_in_tracts.head())

print(f"\nTotal points: {len(sample_points)}")
print(f"Points joined to a census tract: {points_in_tracts['geoid10'].count()}")
print(f"Points not joined to any census tract (outside Philadelphia or gaps): {len(sample_points) - points_in_tracts['geoid10'].count()}")

print("\nThis demonstrates that GeoPandas can successfully perform a point-in-polygon join.")
````

````python
# Rename columns in philly_heat_gdf for easier merging (match crosswalk 'TRACT')
philly_heat_gdf_for_merge = philly_heat_gdf.rename(columns={'geoid10': 'TRACT'})

# Ensure 'TRACT' is string type in philly_heat_gdf_for_merge
philly_heat_gdf_for_merge['TRACT'] = philly_heat_gdf_for_merge['TRACT'].astype(str)

# Create a consistent 'TRACT' column in philly_tract_zcta_crosswalk by combining STATE, COUNTY, and TRACT
# Ensure leading zeros for STATE (2 digits), COUNTY (3 digits), and TRACT (6 digits)
philly_tract_zcta_crosswalk['TRACT'] = \
    philly_tract_zcta_crosswalk['STATE'].astype(str).str.zfill(2) + \
    philly_tract_zcta_crosswalk['COUNTY'].astype(str).str.zfill(3) + \
    philly_tract_zcta_crosswalk['TRACT'].astype(str).str.zfill(6)


# Diagnostic: Print unique TRACT values from both DataFrames to check for mismatches
print("\nUnique TRACT values in philly_heat_gdf_for_merge (first 10):\n", philly_heat_gdf_for_merge['TRACT'].unique()[:10])
print("\nUnique TRACT values in philly_tract_zcta_crosswalk (first 10):\n", philly_tract_zcta_crosswalk['TRACT'].unique()[:10])

# Get the intersection of TRACT values
common_tracts = set(philly_heat_gdf_for_merge['TRACT']).intersection(set(philly_tract_zcta_crosswalk['TRACT']))
print(f"\nNumber of common TRACTs before merge: {len(common_tracts)}")

# Merge the Philadelphia heat vulnerability data with the crosswalk file
# This will link each census tract to its corresponding ZCTA
philly_merged_ct_zcta = philly_heat_gdf_for_merge.merge(
    philly_tract_zcta_crosswalk,
    on='TRACT',
    how='inner'
)

print("Merged Philadelphia heat vulnerability data with ZCTA crosswalk.")
print("\nFirst 5 rows of the merged data:")
display(philly_merged_ct_zcta.head())
print("\nMerged Data Info:")
philly_merged_ct_zcta.info()
````

````python
# Define the vulnerability score columns
vulnerability_scores = ['hsi_score', 'hei_score', 'hvi_score']

# Identify the weight column for aggregation. 'POPPT' is population for the tract portion.
# 'AREAPT' is area for the tract portion. Let's use POPPT for a population-weighted average.
weight_col = 'POPPT'

# Calculate population-weighted average for each ZCTA
philly_zcta_vulnerability = philly_merged_ct_zcta.groupby('ZCTA5').apply(
    lambda x:
        pd.Series({
            score: (x[score] * x[weight_col]).sum() / x[weight_col].sum()
            for score in vulnerability_scores
        })
).reset_index()

# Rename the ZCTA5 column to 'zcta' to match the existing df_clean
philly_zcta_vulnerability = philly_zcta_vulnerability.rename(columns={'ZCTA5': 'zcta'})

print("Aggregated Philadelphia heat vulnerability scores to ZCTA level.")
print("\nFirst 5 rows of aggregated ZCTA data:")
display(philly_zcta_vulnerability.head())
print("\nAggregated Data Info:")
philly_zcta_vulnerability.info()

print(f"\nNumber of Philadelphia ZCTAs with new vulnerability data: {len(philly_zcta_vulnerability)}")
````

````python
# Merge the aggregated Philadelphia ZCTA vulnerability data with the main DataFrame
# We will perform a left merge on 'df_clean' to retain all ZCTAs in the national dataset
# and add the Philadelphia-specific heat vulnerability scores where available.

df_clean = df_clean.merge(philly_zcta_vulnerability, on='zcta', how='left')

print("Merged Philadelphia ZCTA vulnerability data into df_clean.")
print("\nFirst 5 rows of df_clean with new vulnerability scores:")
display(df_clean.head())
print("\nUpdated df_clean Info:")
df_clean.info()

# Diagnostic: Check how many Philadelphia ZCTAs now have non-null heat vulnerability scores
philly_zcta_hvi_check = df_clean[df_clean['zcta'].isin(PHILLY_ZCTAS)]['hvi_score'].count()
print(f"\nNumber of Philadelphia ZCTAs in df_clean with 'hvi_score': {philly_zcta_hvi_check}")
````

````python


# Full PLACES download — filtered to 8 measures, CrudePrev only
# Expected: ~30-90 seconds

TARGET_MEASURES = ['COPD', 'CHD', 'DIABETES', 'OBESITY', 'BPHIGH', 'GHLTH', 'CSMOKING', 'STROKE']

# The CDC PLACES API (kee5-23sr.json) returns data in a wide format where measures and value types
# are combined into column names (e.g., 'COPD_crudeprev').
# We need to explicitly list these columns for selection.
MEASURE_CRUDEPREV_COLS = [f'{m.lower()}_crudeprev' for m in TARGET_MEASURES]

# Define the columns to select from the API directly.
# 'zcta5' for ZCTA, 'totalpopulation' for population, and the crude prevalence for each target measure.
SELECT_COLS = [COL_ZCTA, COL_POP] + MEASURE_CRUDEPREV_COLS

# Construct the $select parameter for the API call.
# We join the column names to get a comma-separated string.
select_clause = ",".join(SELECT_COLS)

# For this wide-format API, we don't use $where clauses on measureid/datavaluetypeid directly.
# We select the specific crude prevalence columns.
# We also don't have a direct 'stateabbr' column in this dataset, so we omit it for now.

params = {
    "$select": select_clause,
    "$limit": 500000, # Increased limit to ensure all ZCTAs are fetched
    "$order": COL_ZCTA
}

print("Downloading CDC PLACES data...")
print("Expected time: 30-90 seconds\n")

resp = requests.get(PLACES_BASE_URL, params=params, timeout=180)
resp.raise_for_status()

places_raw = pd.DataFrame(resp.json())

# Diagnostic: Check for Philly ZCTAs in places_raw (before any further processing)
philly_zcta_check_places_raw = places_raw[COL_ZCTA].isin(PHILLY_ZCTAS).sum()
print(f"Diagnostic: Number of Philly ZCTAs found in places_raw (from API): {philly_zcta_check_places_raw}")

# Now, melt the DataFrame to convert from wide format to long format
# as expected by downstream processing.
places_long = places_raw.melt(
    id_vars=[COL_ZCTA, COL_POP],
    value_vars=MEASURE_CRUDEPREV_COLS,
    var_name='measureid_crudeprev',
    value_name='datavalue'
)

# Extract the actual measureid from the melted 'measureid_crudeprev' column
places_long['measureid'] = places_long['measureid_crudeprev'].str.split('_').str[0].str.upper()
places_long = places_long.drop(columns=['measureid_crudeprev'])

# Add 'CrudePrev' as the datavalue type, as all selected measures are crude prevalence
places_long['datavaluetype'] = 'CrudePrev'

# Rename columns to match the expected schema of the rest of the notebook
places_long = places_long.rename(columns={
    COL_ZCTA: 'locationid',
    COL_POP: 'totalpopulation'
})

# Add a placeholder for 'stateabbr' as it's used later, even if not directly from this API
# In a real scenario, stateabbr would be merged from a separate ZCTA-to-State mapping
places_long['stateabbr'] = 'PA' # Placeholder, will need to be properly joined later if needed

places_long['datavalue'] = pd.to_numeric(places_long['datavalue'], errors='coerce')
places_long['totalpopulation'] = pd.to_numeric(places_long['totalpopulation'], errors='coerce')
places_long['locationid'] = places_long['locationid'].astype(str).str.zfill(5)

print(f"Downloaded {len(places_long):,} rows")
print(f" Unique ZCTAs: {places_long['locationid'].nunique():,}")
print(f" Measures found: {sorted(places_long['measureid'].unique().tolist())}")
print(f" Data value types found: {places_long['datavaluetype'].unique().tolist()}")
print(f"\nMissing values:")
print(places_long[['datavalue', 'totalpopulation']].isnull().sum())

# Diagnostic: Check for Philly ZCTAs in places_long
philly_zcta_check_places_long = places_long['locationid'].isin(PHILLY_ZCTAS).sum()
print(f"\nDiagnostic: Number of Philly ZCTAs found in places_long: {philly_zcta_check_places_long}")
````

````python

# Calculate rate features from raw ACS counts
# All rates expressed as percentages (0-100)

acs = acs_df.copy()

acs['poverty_rate'] = np.where(
    acs['poverty_universe'] > 0,
    acs['below_poverty'] / acs['poverty_universe'] * 100,
    np.nan
)

acs['renter_rate'] = np.where(
    acs['tenure_universe'] > 0,
    acs['renter_occupied'] / acs['tenure_universe'] * 100,
    np.nan
)

pre1980 = (
    acs['built_1970_1979'] + acs['built_1960_1969'] +
    acs['built_1950_1959'] + acs['built_1940_1949'] +
    acs['built_1939_earlier']
)
acs['pct_housing_pre1980'] = np.where(
    acs['housing_unit_universe'] > 0,
    pre1980 / acs['housing_unit_universe'] * 100,
    np.nan
)

elderly = (
    acs['male_65_66'] + acs['male_67_69'] + acs['male_70_74'] +
    acs['male_75_79'] + acs['male_80_84'] + acs['male_85_plus'] +
    acs['female_65_66'] + acs['female_67_69'] + acs['female_70_74'] +
    acs['female_75_79'] + acs['female_80_84'] + acs['female_85_plus']
)
acs['pct_elderly'] = np.where(
    acs['total_population'] > 0,
    elderly / acs['total_population'] * 100,
    np.nan
)

disability = (
    acs['male_u5_disability'] + acs['male_5_17_disability'] +
    acs['male_18_34_disability'] + acs['male_35_64_disability'] +
    acs['male_65_74_disability'] + acs['male_75plus_disability'] +
    acs['female_u5_disability'] + acs['female_5_17_disability'] +
    acs['female_18_34_disability'] + acs['female_35_64_disability'] +
    acs['female_65_74_disability'] + acs['female_75plus_disability']
)
acs['disability_rate'] = np.where(
    acs['disability_universe'] > 0,
    disability / acs['disability_universe'] * 100,
    np.nan
)

acs_features = acs[[
    'zcta', 'total_population',
    'poverty_rate', 'renter_rate', 'pct_housing_pre1980',
    'pct_elderly', 'disability_rate'
]].copy()

acs_features = acs_features[acs_features['total_population'] > 0].copy()

print(f"ACS features engineered: {len(acs_features):,} populated ZCTAs")
print(f"\nNational feature summary (%):")
feat_cols = ['poverty_rate', 'renter_rate', 'pct_housing_pre1980', 'pct_elderly', 'disability_rate']
print(acs_features[feat_cols].describe().round(2).to_string())

# Diagnostic: Check for Philly ZCTAs in acs_features
philly_zcta_check_acs_features = acs_features['zcta'].isin(PHILLY_ZCTAS).sum()
print(f"\nDiagnostic: Number of Philly ZCTAs found in acs_features: {philly_zcta_check_acs_features}")
````

````python
# Composite Structural Vulnerability Index

The CDC Heat and Health Index (HHI) — the current baseline for heat risk assessment — produces
a structural vulnerability score per ZCTA from demographic and health data alone.

We build an equivalent Composite Structural Vulnerability Index here. This serves as:
1. A baseline comparator: SustAInable must outperform it (target AUC ≥ 0.85 vs. HHI baseline ~0.72–0.74)
2. Proof that the feature layer captures real vulnerability signal
3. A standalone useful artifact for cities that cannot yet integrate weather forecast data

**Method:** Min-max normalize each feature → weighted average → scale to 0–100

The vulnerability index will be computed in the next step to ensure it includes the recently merged Philadelphia-specific heat vulnerability scores.
````

````python
import numpy as np

# Composite Structural Vulnerability Index Calculation

# Make a copy to avoid modifying the original df_clean during feature engineering for the index
idx_df = df_clean.copy()

# Define the features and their weights for the Composite Structural Vulnerability Index
# These weights are derived from the kernel state FEATURE_WEIGHTS variable
FEATURE_WEIGHTS = {
    'pct_hypertension': 0.1,
    'pct_chd': 0.08,
    'pct_copd': 0.08,
    'pct_smoking': 0.04,
    'pct_diabetes': 0.08,
    'pct_poor_health': 0.08,
    'pct_obesity': 0.05,
    'pct_stroke': 0.07,
    'poverty_rate': 0.12,
    'renter_rate': 0.07,
    'pct_housing_pre1980': 0.08,
    'pct_elderly': 0.1,
    'disability_rate': 0.05
}

index_features = list(FEATURE_WEIGHTS.keys())

# --- Step 1: Integrate Philadelphia heat vulnerability scores into health features ---
# Identify Philadelphia ZCTAs
philly_zcta_mask = idx_df['zcta'].isin(PHILLY_ZCTAS)

# Handle cases where hvi_score might be NaN for some Philly ZCTAs
# This should ideally not happen after the previous merge if data is clean.
if idx_df[philly_zcta_mask]['hvi_score'].isnull().any():
    print("Warning: Some Philadelphia ZCTAs have missing hvi_score values. Imputing with median for scaling.")
    # For robust scaling, fill any individual NaNs within Philly hvi_scores temporarily
    philly_hvi_temp = idx_df.loc[philly_zcta_mask, 'hvi_score'].fillna(idx_df.loc[philly_zcta_mask, 'hvi_score'].median())
else:
    philly_hvi_temp = idx_df.loc[philly_zcta_mask, 'hvi_score']

# Scale hvi_score for Philadelphia ZCTAs to a 0-100 range
if not philly_hvi_temp.empty and not pd.isna(philly_hvi_temp).all():
    min_hvi = philly_hvi_temp.min()
    max_hvi = philly_hvi_temp.max()

    if max_hvi == min_hvi: # Avoid division by zero if all hvi_scores are the same
        idx_df.loc[philly_zcta_mask, 'scaled_philly_hvi_score'] = 50.0 # Assign a mid-range value
    else:
        idx_df.loc[philly_zcta_mask, 'scaled_philly_hvi_score'] = (
            ((philly_hvi_temp - min_hvi) / (max_hvi - min_hvi)) * 100
        )
else:
    print("No valid hvi_score data for Philadelphia ZCTAs to scale or all were NaN. Filling scaled_philly_hvi_score with NaN for these.")
    idx_df.loc[philly_zcta_mask, 'scaled_philly_hvi_score'] = np.nan # Ensure column exists if no valid scores

# Identify health features for imputation (these are NaN for Philly ZCTAs from CDC PLACES)
health_features = [
    'pct_hypertension', 'pct_chd', 'pct_copd', 'pct_smoking',
    'pct_diabetes', 'pct_poor_health', 'pct_obesity', 'pct_stroke'
]

# For Philadelphia ZCTAs, fill the NaN health features with the scaled hvi_score
# Only fill if the original value is NaN (which it should be for Philly for these columns)
for col in health_features:
    idx_df.loc[philly_zcta_mask & idx_df[col].isnull(), col] = idx_df.loc[philly_zcta_mask & idx_df[col].isnull(), 'scaled_philly_hvi_score']

# Drop the temporary scaled_philly_hvi_score column as it's been used for imputation
idx_df = idx_df.drop(columns=['scaled_philly_hvi_score'], errors='ignore')

# --- Step 2: Impute any remaining NaNs (for national ZCTAs that might have few missing values) ---
# For national ZCTAs, a small number of NaNs might remain after the initial merge.
# We will impute these with the national median for each feature.
for feature in index_features:
    if idx_df[feature].isnull().any():
        median_val = idx_df[feature].median()
        idx_df[feature] = idx_df[feature].fillna(median_val)
        print(f"Imputed {idx_df[feature].isnull().sum()} NaNs in '{feature}' with median {median_val:.2f}")

# Drop ZCTAs if they still have NaNs in critical index features after imputation (should be rare now)
idx_df = idx_df.dropna(subset=index_features).copy()
print(f"Number of ZCTAs after NaN imputation and final drop: {len(idx_df):,} ")

# --- Step 3: Min-Max Normalize features ---
# Normalize all index features to a 0-1 range
normalized_features = idx_df[index_features].copy()
for col in index_features:
    min_val = normalized_features[col].min()
    max_val = normalized_features[col].max()
    if (max_val - min_val) > 0:
        normalized_features[col] = (normalized_features[col] - min_val) / (max_val - min_val)
    else:
        normalized_features[col] = 0.0 # If all values are the same, normalize to 0

# --- Step 4: Calculate Weighted Average ---
# Apply the weights to the normalized features
weighted_sum = sum(
    normalized_features[col] * FEATURE_WEIGHTS[col]
    for col in index_features
)

# --- Step 5: Scale the result to 0-100 ---
# The weighted_sum is already a value between 0 and 1. Scale it to 0-100.
idx_df['vulnerability_index'] = weighted_sum * 100

# --- Step 6: Assign Vulnerability Tiers ---
# Define tiers based on quantiles. Example: Low (bottom 70%), Moderate (next 29%), High (top 1%)
low_threshold = idx_df['vulnerability_index'].quantile(0.70)
moderate_threshold = idx_df['vulnerability_index'].quantile(0.99) # Top 1% as High

def assign_tier(score):
    if score >= moderate_threshold:
        return 'High'
    elif score >= low_threshold:
        return 'Moderate'
    else:
        return 'Low'

idx_df['vulnerability_tier'] = idx_df['vulnerability_index'].apply(assign_tier)

print(f"\nVulnerability index computed: {len(idx_df):,} ZCTAs")
print("\nNational distribution (score 0-100):")
print(idx_df['vulnerability_index'].describe().to_string())

print("\nTier breakdown:")
tc = idx_df['vulnerability_tier'].value_counts()
for tier in ['Low', 'Moderate', 'High']:
    if tier in tc:
        pct = tc[tier] / len(idx_df) * 100
        print(f" {tier:<9}: {tc[tier]:>6,} ZCTAs ({pct:.1f}%)")
    else:
        print(f" {tier:<9}: No ZCTAs in this tier.")
````
